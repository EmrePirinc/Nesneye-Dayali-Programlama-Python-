## try except


````python
try:
    sayi1=int(input("sayi1 giriniz"))
    sayi2=int(input("sayi2 giriniz"))
    sonuc=sayi1/sayi2
    print(sonuc)
except:                        #Tüm hataları *genelledik* burada. Herhangibir hata olduğunda bu blok çalışır.
    print("hata oluştu")

#Eğer kullanıcı tarafından hatalı giriş olursa try bloğu içerisindeki kod çalışmaz .except bloğu çalışır bu durumda.
````

**Hatalı Giriş Örneği:**

Kullanıcı sayi2 için "0" girerse, sıfıra bölme hatası oluşur ve program çöker. Ancak ```try-except``` bloğu sayesinde program çökmez ve kullanıcıya hata mesajı gösterilir.

**Neden ```try-except``` Kullanmalıyız?**  

**Kullanıcı Dostu Hata Mesajları:** Son kullanıcılar programlama hatalarını anlamayabilir. ```try-except``` ile onlara daha açıklayıcı mesajlar gösterebiliriz.

**Programın Kararlı Çalışması:** Hataların programı çökertmesini önler ve programın kararlı bir şekilde çalışmasını sağlar.

**Hata Ayıklama:** Hataların nerede ve nasıl oluştuğunu belirlemeye yardımcı olur.

**Daha Gelişmiş Hata Yakalama:**
Farklı hata türlerini yakalamak için birden fazla ```except bloğu``` kullanılabilir.  

```finally``` bloğu ile hata olsun veya olmasın mutlaka yapılması gereken işlemler belirtilebilir.


````python

try:
    sayi1=int(input("sayi1 giriniz")) #1
    sayi2=int(input("sayi2 giriniz")) #a
    sonuc=sayi1/sayi2
    print(sonuc)
except ValueError as e:
    print(str(e)) #print("hata oluştu")

#invalid literal for int() with base 10: 'a' program onluk tabanda bir sayi ifadesi girmeni bekliyordu
# a ifadesi girdiğimiz için bu hatayı döndürür.
````

````python

try:
    sayi1=int(input("sayi1 giriniz")) #5
    sayi2=int(input("sayi2 giriniz")) #0
    sonuc=sayi1/sayi2
    print(sonuc)
except ValueError as e:
    print(str(e))


#biz sadece valueError "a" hata dediğimiz için 0 girdiğimizde bunu hata olarak kabul etmez ve kod da gene hata alırız.

#Tüm hataları genellemeyip sadece değer hatasına(ValueError) hata dedik.  bunun dışında bir hata olursa program gene kırılacaktır.
#5 ve 0 girdiğimiz de program kırılır ve sıfıra bölünme hatası verir . (ZeroDivisionError)
````

**Bu sorunu 2 şekilde çözebiliriz**

1) ya except bloğunu **genel** bir hataya çeviririz.
 (  except: ) Sadece bloğu açıp özel hata belirtmeyerek.  

ya da  

2) farklı bir except bloğu daha açarak oraya sıfıra bölünme hatasını da tanımlarız.

````python
try:
    sayi1=int(input("sayi1 giriniz"))
    sayi2=int(input("sayi2 giriniz"))
    sonuc=sayi1/sayi2
    print(sonuc)
except ValueError as e:
    print("değer girişinde hata oluştu")
except ZeroDivisionError as e:                    #farklı bir except bloğu daha açarak oraya sıfıra bölünme hatasını da tanımladık
    print("bölünen sayi 0 olamaz")
````

````python
try:
    sayi1=int(input("sayi1 giriniz"))
    sayi2=int(input("sayi2 giriniz"))
    sonuc=sayi1/sayi2
    print(sonuc)
except ValueError as e:
    print("değer girişinde hata oluştu")
except ZeroDivisionError as e:
    print("bölünen sayi 0 olamaz")
finally:
    print("işlem bitti")         #hata olsada olmasada bu blok her halükarda çalışır

````

### Python'da Dosya Okuma Yöntemleri

**1. `read()` Fonksiyonu**

  - `read()` fonksiyonu, dosyanın tamamını okur. Eğer bir parametre verilirse, belirtilen miktarda karakter okur.
  - `read()` kullanıldığında, dosyanın tamamını bir tek string (metin) olarak döndürür.

**2. `readline()` Fonksiyonu**

  - `readline()` fonksiyonu, dosyadaki imlecin bulunduğu satırı okur. Fonksiyon her çağrıldığında imleci bir sonraki satıra taşır.
  - İlk `readline()` çağrısı ilk satırı, ikinci çağrı ikinci satırı okur. Bu şekilde, her çağrı sonraki satırı okur ve imleci bir sonraki satıra taşır.

**3. `readlines()` Fonksiyonu**

  - `readlines()` fonksiyonu, dosyanın tüm satırlarını okuyup, her bir satırı bir Python listesinin elemanı olarak döndürür.
  - Bu liste üzerinde döngülerle işlem yapılabilir. Örneğin, dosyanın her bir satırı listesinin bir elemanıdır ve her elemanın sonunda genellikle bir satır sonu karakteri (`\n`) bulunur.

**Özet:**
- `read()` dosyanın tamamını bir string olarak okur.
- `readline()` dosyadaki imlecin bulunduğu satırı okur ve sonraki satıra geçer.
- `readlines()` tüm satırları bir liste olarak döndürür, her satır listesinin bir elemanıdır.

## `a+` ve `r+` Kipleri Arasındaki Farklar

| **Özellik**                 | **`a+`**                  | **`r+`**                 |
|-----------------------|--------------------|--------------------|
| Dosya mevcut değilse   | Yeni dosya oluşturur | Hata verir           |
| Dosya işaretçisi       | Dosyanın sonu       | Dosyanın başı        |
| Okuma                  | Evet                   | Evet                   |
| Yazma                  | Evet (sona ekler)     | Evet (baştan yazar)   |
| Mevcut içeriği korur mu? | Evet                   | Hayır (üzerine yazar) |

**Açıklama:**

* **Dosya mevcut değilse:** Her iki kip de dosya mevcut değilse yeni bir dosya oluşturur. Ancak `r+` kipinde, dosya mevcut değilse hata verir.
* **Dosya işaretçisi:** `a+` kipinde, dosya işaretçisi dosyanın sonuna konumlandırılır. Bu, yazılan verilerin mevcut içeriğin sonuna ekleneceği anlamına gelir. `r+` kipinde ise, dosya işaretçisi dosyanın başına konumlandırılır. Bu, yazılan verilerin mevcut içeriğin üzerine yazılacağı anlamına gelir.
* **Okuma:** Her iki kip de dosyadan veri okumaya izin verir.
* **Yazma:** Her iki kip de dosyaya veri yazmaya izin verir. Ancak `a+` kipinde yazma işlemi dosyanın sonuna ekleme yaparken, `r+` kipinde yazma işlemi dosyanın başından başlar ve mevcut içeriğin üzerine yazar.
* **Mevcut içeriği korur mu?:** `a+` kipi mevcut dosya içeriğini korur ve yazılan verileri sonuna ekler. `r+` kipi ise mevcut dosya içeriğini korumaz ve yazılan verilerle üzerine yazar.

**Hangi Kipi Kullanmalıyım?**

* **Mevcut dosya içeriğini korumak ve sonuna veri eklemek istiyorsanız:** `a+` kipini kullanın. Örneğin, bir günlük dosyasına yeni girdiler eklemek için bu kipi kullanabilirsiniz.
* **Mevcut dosya içeriğini değiştirmek ve baştan veri yazmak istiyorsanız:** `r+` kipini kullanın. Örneğin, bir dosyanın belirli bir bölümünü güncellemek için bu kipi kullanabilirsiniz.
